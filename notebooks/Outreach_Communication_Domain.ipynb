# Outreach & Communication Domain

**Purpose:** Public engagement, awareness, and impact reporting.

**Tables:** `social_media_posts`, `safehouse_monthly_metrics`, `public_impact_snapshots`

```python
from pathlib import Path

import pandas as pd

_csv = Path("notebooks/lighthouse_csv_v7")
DATA_DIR = _csv if _csv.is_dir() else Path("lighthouse_csv_v7")
if not DATA_DIR.is_dir():
    raise FileNotFoundError(
        f"Data folder not found: {DATA_DIR.resolve()} — run from repo root or the notebooks folder."
    )

TABLES = [
    "social_media_posts",
    "safehouse_monthly_metrics",
    "public_impact_snapshots",
]

dataframes = {name: pd.read_csv(DATA_DIR / f"{name}.csv") for name in TABLES}

social_media_posts = dataframes["social_media_posts"]
safehouse_monthly_metrics = dataframes["safehouse_monthly_metrics"]
public_impact_snapshots = dataframes["public_impact_snapshots"]
```
