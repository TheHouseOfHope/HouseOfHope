# Donor & Support Domain

**Purpose:** Funding, partnerships, and resource flow into the organization.

**Tables:** `safehouses`, `partners`, `partner_assignments`, `supporters`, `donations`, `in_kind_donation_items`, `donation_allocations`

```python
from pathlib import Path

import pandas as pd

DATA_DIR = Path("lighthouse_csv_v7")
if not DATA_DIR.is_dir():
    raise FileNotFoundError(
        f"Data folder not found: {DATA_DIR.resolve()} — run this notebook from the INTEXII project root."
    )

TABLES = [
    "safehouses",
    "partners",
    "partner_assignments",
    "supporters",
    "donations",
    "in_kind_donation_items",
    "donation_allocations",
]

dataframes = {name: pd.read_csv(DATA_DIR / f"{name}.csv") for name in TABLES}

safehouses = dataframes["safehouses"]
partners = dataframes["partners"]
partner_assignments = dataframes["partner_assignments"]
supporters = dataframes["supporters"]
donations = dataframes["donations"]
in_kind_donation_items = dataframes["in_kind_donation_items"]
donation_allocations = dataframes["donation_allocations"]
```
